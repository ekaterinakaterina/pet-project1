```python
def get_channel_stats(youtube, channel_ids):
    """
    Получение статистики канала: название, количество подписчиков, количество просмотров, количество видео, загрузить плейлист
    
    Параметры:    
    youtube: объект сборки из googleapiclient.discovery
    channels_ids: список идентификаторов каналов
    
    Возвращает:
    Data Frame, содержащий статистику канала для всех каналов в предоставленном списке: название, количество подписчиков,
    количество просмотров, количество видео, загрузки пл
    
    """
    all_data = []
    request = youtube.channels().list(
                part='snippet,contentDetails,statistics',
                id=','.join(channel_ids))
    response = request.execute() 
    
    for i in range(len(response['items'])):
        data = dict(channelName = response['items'][i]['snippet']['title'],
                    subscribers = response['items'][i]['statistics']['subscriberCount'],
                    views = response['items'][i]['statistics']['viewCount'],
                    totalVideos = response['items'][i]['statistics']['videoCount'],
                    playlistId = response['items'][i]['contentDetails']['relatedPlaylists']['uploads'])
        all_data.append(data)
    
    return pd.DataFrame(all_data)

def get_video_ids(youtube, playlist_id):
    """
    Получение списка идентификаторов всех видео в данном плейлисте
    
    Параметры:
    youtube: объект сборки из googleapiclient.discovery
    playlist_id: идентификатор плейлиста канала
    
    Возвращает:
    Список идентификаторов всех видео в плейлисте
    
    """
    
    request = youtube.playlistItems().list(
                part='contentDetails',
                playlistId = playlist_id,
                maxResults = 50)
    response = request.execute()
    
    video_ids = []
    
    for i in range(len(response['items'])):
        video_ids.append(response['items'][i]['contentDetails']['videoId'])
        
    next_page_token = response.get('nextPageToken')
    more_pages = True
    
    while more_pages:
        if next_page_token is None:
            more_pages = False
        else:
            request = youtube.playlistItems().list(
                        part='contentDetails',
                        playlistId = playlist_id,
                        maxResults = 50,
                        pageToken = next_page_token)
            response = request.execute()
    
            for i in range(len(response['items'])):
                video_ids.append(response['items'][i]['contentDetails']['videoId'])
            
            next_page_token = response.get('nextPageToken')
        
    return video_ids

def get_video_details(youtube, video_ids):
    """
    Получить видеостатистику всех видео с заданными идентификаторами
    
    Параметры:    
    youtube: объект сборки из googleapiclient.discovery
    video_ids: список идентификаторов видео
    
    Возвращает:
    Датафрейм со статистикой видео, т.е.:
        заголовок канала, название, описание, теги, когда опубликовано,
        счетчик просмотров, счетчик лайков, 'favoriteCount', счетчик комментариев,
        длительность, определение, заголовок.
    """
        
    all_video_info = []
    
    for i in range(0, len(video_ids), 50):
        request = youtube.videos().list(
            part="snippet,contentDetails,statistics",
            id=','.join(video_ids[i:i+50])
        )
        response = request.execute() 

        for video in response['items']:
            stats_to_keep = {'snippet': ['channelTitle', 'title', 'description', 'tags', 'publishedAt'],
                             'statistics': ['viewCount', 'likeCount', 'favouriteCount', 'commentCount'],
                             'contentDetails': ['duration', 'definition', 'caption']
                            }
            video_info = {}
            video_info['video_id'] = video['id']

            for k in stats_to_keep.keys():
                for v in stats_to_keep[k]:
                    try:
                        video_info[v] = video[k][v]
                    except:
                        video_info[v] = None

            all_video_info.append(video_info)
            
    return pd.DataFrame(all_video_info)

def get_comments_in_videos(youtube, video_ids):
    """
    Получить популярные комментарии в виде текста из всех видео с заданными идентификаторами (только первые 10 комментариев
    из-за ограничений API Youtube)
    
    Параметры:    
    youtube: объект сборки из googleapiclient.discovery
    video_ids: список идентификаторов видео
    
    Возвращает:
    Фрейм данных с идентификаторами видео и комментариями в текстовом виде.
    
    """
    all_comments = []
    
    for video_id in video_ids:
        try:   
            request = youtube.commentThreads().list(
                part="snippet,replies",
                videoId=video_id
            )
            response = request.execute()
        
            comments_in_video = [comment['snippet']['topLevelComment']['snippet']['textOriginal'] for comment in response['items'][0:10]]
            comments_in_video_info = {'video_id': video_id, 'comments': comments_in_video}

            all_comments.append(comments_in_video_info)
            
        except: 
            # When error occurs - most likely because comments are disabled on a video
            print('Could not get comments for video ' + video_id)
        
    return pd.DataFrame(all_comments)     
```

# Обогащение данных

Я хочу обогатить данные для дальнейшего анализа, например:
   - создать столбец даты публикации с другим столбцом, показывающим день недели, когда было опубликовано видео, что будет полезно для последующего анализа.

```python
# Добавить количество тегов
video_df['tagsCount'] = video_df['tags'].apply(lambda x: 0 if x is None else len(x))
```

```python
# Комментарии и лайки на 1000 просмотров
video_df['likeRatio'] = video_df['likeCount']/ video_df['viewCount'] * 1000
video_df['commentRatio'] = video_df['commentCount']/ video_df['viewCount'] * 1000
```

Поскольку мне интересно посмотреть, о чем создатели снимают видео и какие термины чаще всего встречаются в названиях их видео, я создам облако слов для наиболее распространенных слов. Сначала нам нужно удалить стоп-слова, такие как «ты», «я», «тот» и т. д., которые часто используются в заголовках.

```python
stop_words = list(stopwords.words('russian'))

video_df['title_no_stopwords'] = video_df['title'].apply(lambda x: [item for item in str(x).split() if item not in stop_words])

all_words = list([a for b in video_df['title_no_stopwords'].tolist() for a in b])
all_words_str = ' '.join(all_words) 
```

```python
def plot_cloud(wordcloud):
    plt.figure(figsize=(30, 20))
    plt.imshow(wordcloud) 
    plt.axis("off");

wordcloud = WordCloud(width = 2000, height = 1000, random_state=1, background_color='black', 
                      colormap='viridis', collocations=False).generate(all_words_str)
plot_cloud(wordcloud)
```
